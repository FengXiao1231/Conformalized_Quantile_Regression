```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, KFold
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats.mstats import winsorize

# Set random seed
np.random.seed(42)
tf.random.set_seed(42)

# ========== Parameter Settings ==========
alpha = 0.1  # Corresponding to 90% coverage
alpha_l = alpha / 2
alpha_u = 1 - alpha_l

df = pd.read_csv('loan_sampled.csv')
df = winsorize(df, limits=[0.01, 0.01])
# Load data (please modify according to your actual data path)
url = r'F:\Stony Brook University\2024 Fall\AMS518\Project\Real estate valuation data set.xlsx'
df = pd.read_excel(url)

# Data preprocessing
df.columns = ['No', 'Transaction_Date', 'House_Age', 'Distance_To_MRT', 'Number_Convenience_Stores', 'Latitude', 'Longitude', 'House_Price']
df.drop(['No', 'Transaction_Date'], axis=1, inplace=True)

X = df.drop('House_Price', axis=1)
y = df['House_Price']

###################################
# Define necessary functions (same as previous code)
###################################
def coverage_and_length(lower_preds, upper_preds, y_true):
    coverage = np.mean((y_true >= lower_preds) & (y_true <= upper_preds))
    avg_length = np.mean(upper_preds - lower_preds)
    return coverage, avg_length

def pinball_loss(alpha_q):
    def loss(y_true, y_pred):
        diff = y_true - y_pred
        return tf.reduce_mean(tf.maximum(alpha_q*diff, (alpha_q-1)*diff))
    return loss

# Mean model training function
def train_ridge(X, y):
    ridge = Ridge(alpha=1.0)
    ridge.fit(X, y)
    return ridge

def train_rf_mean(X, y):
    rf = RandomForestRegressor(n_estimators=1000, random_state=42)
    rf.fit(X, y)
    return rf

def build_nn_mean(input_dim):
    model = models.Sequential([
        layers.Dense(64, activation='relu', input_shape=(input_dim,)),
        layers.Dense(64, activation='relu'),
        layers.Dense(1)
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(5e-4), loss='mse')
    return model

def train_nn_mean(X, y):
    model = build_nn_mean(X.shape[1])
    es = tf.keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)
    model.fit(X, y, epochs=1000, batch_size=64, validation_split=0.1, verbose=0, callbacks=[es])
    return model

# Quantile model training function (GBR)
def train_quantile_gbr(X, y, alpha_q):
    gbr = GradientBoostingRegressor(loss='quantile', alpha=alpha_q,
                                    n_estimators=1000, max_depth=5,
                                    learning_rate=0.05, random_state=42)
    gbr.fit(X, y)
    return gbr

# QNN
def build_nn_quantile(input_dim):
    model = models.Sequential([
        layers.Dense(64, activation='relu', input_shape=(input_dim,)),
        layers.Dense(64, activation='relu'),
        layers.Dense(1)
    ])
    return model

def train_qnn(X, y, alpha_q):
    model = build_nn_quantile(X.shape[1])
    model.compile(optimizer=tf.keras.optimizers.Adam(5e-4), loss=pinball_loss(alpha_q))
    es = tf.keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)
    model.fit(X, y, epochs=1000, batch_size=64, validation_split=0.1, verbose=0, callbacks=[es])
    return model

def train_qnn_pair(X, y, alpha_l, alpha_u):
    model_lower = train_qnn(X, y, alpha_l)
    model_upper = train_qnn(X, y, alpha_u)
    return model_lower, model_upper

# SCP function
def split_conformal_calibrate(f, X_cal, y_cal, alpha=0.1):
    cal_preds = f(X_cal)
    residuals = np.abs(y_cal - cal_preds)
    q_res = np.quantile(residuals, 1 - alpha)
    return q_res

def split_conformal_predict(f, X, q_res):
    preds = f(X)
    return preds - q_res, preds + q_res

# LACP function
def train_mu_sigma_for_lacp(X_train, y_train, mu_model_type='rf'):
    if mu_model_type == 'ridge':
        mu_model = train_ridge(X_train, y_train)
    elif mu_model_type == 'rf':
        mu_model = train_rf_mean(X_train, y_train)
    elif mu_model_type == 'nn':
        mu_model = train_nn_mean(X_train, y_train)
    else:
        raise ValueError("Unknown mu_model_type")

    mu_pred_train = mu_model.predict(X_train)
    mu_pred_train = np.array(mu_pred_train).ravel()
    R_train = np.abs(y_train - mu_pred_train)

    sigma_model = RandomForestRegressor(n_estimators=500, random_state=42)
    sigma_model.fit(X_train, R_train)
    return mu_model, sigma_model

def lacp_calibrate(mu_model, sigma_model, X_cal, y_cal, alpha=0.1):
    mu_pred_cal = mu_model.predict(X_cal)
    sigma_pred_cal = sigma_model.predict(X_cal)
    sigma_pred_cal = np.array(sigma_pred_cal).ravel()
    R_tilde_cal = np.abs(y_cal - mu_pred_cal)/(sigma_pred_cal + 1e-9)
    Q_1_alpha = np.quantile(R_tilde_cal, 1 - alpha)
    return Q_1_alpha

def lacp_predict(mu_model, sigma_model, X, Q_1_alpha):
    mu_pred = mu_model.predict(X)
    sigma_pred = sigma_model.predict(X)
    return mu_pred - sigma_pred*Q_1_alpha, mu_pred + sigma_pred*Q_1_alpha

# CQR function
def cqr_calibrate(q_lower, q_upper, X_cal, y_cal, alpha=0.1):
    cal_lower = q_lower(X_cal)
    cal_upper = q_upper(X_cal)
    E = np.maximum(cal_lower - y_cal, y_cal - cal_upper)
    Q = np.quantile(E, 1 - alpha)
    return Q

def cqr_predict(q_lower, q_upper, X, Q):
    lower_preds = q_lower(X) - Q
    upper_preds = q_upper(X) + Q
    return lower_preds, upper_preds

###################################
# Use KFold for cross-validation
###################################
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# We need to record coverage and length for various methods
methods = [
    'Ridge','Random Forest','Neural Net',
    'Ridge_LACP','RF_LACP','NN_LACP',
    'CQR_QRF','CQR_QNN',
    'QRF_Quantile_NoConformal','QNN_Quantile_NoConformal'
]

coverage_results = {m:[] for m in methods}
length_results = {m:[] for m in methods}

# To display intervals for CQR(GBR) at the last fold
cqr_gbr_interval_lower = None
cqr_gbr_interval_upper = None
y_test_final = None

X_np = X.values if hasattr(X,'values') else X
y_np = y.values if hasattr(y,'values') else y

for fold_id,(train_index, test_index) in enumerate(kf.split(X_np)):
    X_train_full, X_test_fold = X_np[train_index], X_np[test_index]
    y_train_full, y_test_fold = y_np[train_index], y_np[test_index]

    # Split train/calibration sets
    n_cal = int(0.2 * len(X_train_full))
    X_train_fold, X_cal_fold = X_train_full[:-n_cal], X_train_full[-n_cal:]
    y_train_fold, y_cal_fold = y_train_full[:-n_cal], y_train_full[-n_cal:]

    # Train mean models
    ridge = train_ridge(X_train_fold, y_train_fold)
    rf_mean = train_rf_mean(X_train_fold, y_train_fold)
    nn_mean = train_nn_mean(X_train_fold, y_train_fold)

    f_ridge = lambda X: ridge.predict(X)
    f_rf = lambda X: rf_mean.predict(X)
    f_nn = lambda X: nn_mean.predict(X).flatten()

    # Quantile models (GBR Quantile)
    model_lower_gbr = train_quantile_gbr(X_train_fold, y_train_fold, alpha_l)
    model_upper_gbr = train_quantile_gbr(X_train_fold, y_train_fold, alpha_u)
    f_gbr_lower = lambda X: model_lower_gbr.predict(X)
    f_gbr_upper = lambda X: model_upper_gbr.predict(X)

    # QNN
    model_lower_qnn, model_upper_qnn = train_qnn_pair(X_train_fold, y_train_fold, alpha_l, alpha_u)
    f_qnn_lower = lambda X: model_lower_qnn.predict(X).flatten()
    f_qnn_upper = lambda X: model_upper_qnn.predict(X).flatten()

    # ========== SCP ==========
    q_res_ridge_scp = split_conformal_calibrate(f_ridge, X_cal_fold, y_cal_fold, alpha=alpha)
    ridge_lower_scp, ridge_upper_scp = split_conformal_predict(f_ridge, X_test_fold, q_res_ridge_scp)
    cov, length = coverage_and_length(ridge_lower_scp, ridge_upper_scp, y_test_fold)
    coverage_results['Ridge'].append(cov)
    length_results['Ridge'].append(length)

    q_res_rf_scp = split_conformal_calibrate(f_rf, X_cal_fold, y_cal_fold, alpha=alpha)
    rf_lower_scp, rf_upper_scp = split_conformal_predict(f_rf, X_test_fold, q_res_rf_scp)
    cov, length = coverage_and_length(rf_lower_scp, rf_upper_scp, y_test_fold)
    coverage_results['Random Forest'].append(cov)
    length_results['Random Forest'].append(length)

    q_res_nn_scp = split_conformal_calibrate(f_nn, X_cal_fold, y_cal_fold, alpha=alpha)
    nn_lower_scp, nn_upper_scp = split_conformal_predict(f_nn, X_test_fold, q_res_nn_scp)
    cov, length = coverage_and_length(nn_lower_scp, nn_upper_scp, y_test_fold)
    coverage_results['Neural Net'].append(cov)
    length_results['Neural Net'].append(length)

    # ========== LACP ==========
    # Ridge
    mu_ridge, sigma_ridge = train_mu_sigma_for_lacp(X_train_fold, y_train_fold, mu_model_type='ridge')
    Q_ridge_lacp = lacp_calibrate(mu_ridge, sigma_ridge, X_cal_fold, y_cal_fold, alpha=alpha)
    ridge_lower_lacp, ridge_upper_lacp = lacp_predict(mu_ridge, sigma_ridge, X_test_fold, Q_ridge_lacp)
    cov, length = coverage_and_length(ridge_lower_lacp, ridge_upper_lacp, y_test_fold)
    coverage_results['Ridge_LACP'].append(cov)
    length_results['Ridge_LACP'].append(length)

    # RF
    mu_rf, sigma_rf = train_mu_sigma_for_lacp(X_train_fold, y_train_fold, mu_model_type='rf')
    Q_rf_lacp = lacp_calibrate(mu_rf, sigma_rf, X_cal_fold, y_cal_fold, alpha=alpha)
    rf_lower_lacp, rf_upper_lacp = lacp_predict(mu_rf, sigma_rf, X_test_fold, Q_rf_lacp)
    cov, length = coverage_and_length(rf_lower_lacp, rf_upper_lacp, y_test_fold)
    coverage_results['RF_LACP'].append(cov)
    length_results['RF_LACP'].append(length)

    # NN
    mu_nn, sigma_nn = train_mu_sigma_for_lacp(X_train_fold, y_train_fold, mu_model_type='nn')
    Q_nn_lacp = lacp_calibrate(mu_nn, sigma_nn, X_cal_fold, y_cal_fold, alpha=alpha)
    nn_lower_lacp, nn_upper_lacp = lacp_predict(mu_nn, sigma_nn, X_test_fold, Q_nn_lacp)
    cov, length = coverage_and_length(nn_lower_lacp, nn_upper_lacp, y_test_fold)
    coverage_results['NN_LACP'].append(cov)
    length_results['NN_LACP'].append(length)

    # ========== CQR ==========
    # CQR (QRF)
    Q_gbr = cqr_calibrate(f_gbr_lower, f_gbr_upper, X_cal_fold, y_cal_fold, alpha=alpha)
    gbr_lower_cqr, gbr_upper_cqr = cqr_predict(f_gbr_lower, f_gbr_upper, X_test_fold, Q_gbr)
    cov, length = coverage_and_length(gbr_lower_cqr, gbr_upper_cqr, y_test_fold)
    coverage_results['CQR_QRF'].append(cov)
    length_results['CQR_QRF'].append(length)

    # CQR (QNN)
    Q_qnn = cqr_calibrate(f_qnn_lower, f_qnn_upper, X_cal_fold, y_cal_fold, alpha=alpha)
    qnn_lower_cqr, qnn_upper_cqr = cqr_predict(f_qnn_lower, f_qnn_upper, X_test_fold, Q_qnn)
    cov, length = coverage_and_length(qnn_lower_cqr, qnn_upper_cqr, y_test_fold)
    coverage_results['CQR_QNN'].append(cov)
    length_results['CQR_QNN'].append(length)

    # ========== Quantile baseline (No Conformal) ==========
    # QRF
    gbr_lower = f_gbr_lower(X_test_fold)
    gbr_upper = f_gbr_upper(X_test_fold)
    cov, length = coverage_and_length(gbr_lower, gbr_upper, y_test_fold)
    coverage_results['QRF_Quantile_NoConformal'].append(cov)
    length_results['QRF_Quantile_NoConformal'].append(length)

    # QNN
    qnn_lower = f_qnn_lower(X_test_fold)
    qnn_upper = f_qnn_upper(X_test_fold)
    cov, length = coverage_and_length(qnn_lower, qnn_upper, y_test_fold)
    coverage_results['QNN_Quantile_NoConformal'].append(cov)
    length_results['QNN_Quantile_NoConformal'].append(length)

    # Record the last fold's CQR(GBR) results for visualization
    if fold_id == kf.n_splits - 1:
        cqr_gbr_interval_lower = gbr_lower_cqr
        cqr_gbr_interval_upper = gbr_upper_cqr
        y_test_final = y_test_fold

###################################
# Output average results
###################################
print("=== Average Results over 5 folds ===")
for m in methods:
    avg_cov = np.mean(coverage_results[m])
    avg_len = np.mean(length_results[m])
    print(f"{m}: Coverage={avg_cov:.3f}, Length={avg_len:.3f}")

###################################
# Plot comparison of lengths and coverage
###################################
avg_coverages = [np.mean(coverage_results[m]) for m in methods]
avg_lengths = [np.mean(length_results[m]) for m in methods]

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.barplot(x=methods, y=avg_coverages)
plt.xticks(rotation=90)
plt.title("Average Coverage Comparison")
plt.ylabel("Coverage")

plt.subplot(1,2,2)
sns.barplot(x=methods, y=avg_lengths)
plt.xticks(rotation=90)
plt.title("Average Length Comparison")
plt.ylabel("Length")
plt.tight_layout()
plt.show()

###################################
# Display prediction intervals and true values for CQR(GBR)
###################################
# Display a subset of samples (e.g., first 50 samples)
num_samples_to_plot = min(50,len(y_test_final))
x_axis = np.arange(num_samples_to_plot)

plt.figure(figsize=(10,6))
plt.plot(x_axis, y_test_final[:num_samples_to_plot], 'o', color='blue', label='True Values')
mid = (cqr_gbr_interval_lower[:num_samples_to_plot] + cqr_gbr_interval_upper[:num_samples_to_plot]) / 2
plt.plot(x_axis, mid, color='green', label='CQR(QRF) Prediction (Mid)')
plt.fill_between(x_axis, cqr_gbr_interval_lower[:num_samples_to_plot],
                 cqr_gbr_interval_upper[:num_samples_to_plot],
                 color='green', alpha=0.2, label='CQR(QRF) Interval')
plt.xlabel('Test Sample Index')
plt.ylabel('House Price')
plt.title('CQR(QRF) Prediction Intervals vs. True Values')
plt.legend()
plt.show()

###################################
# Use boxplot to show the distribution of different methods under K-fold
###################################

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
coverage_data = [coverage_results[m] for m in methods]
plt.boxplot(coverage_data, labels=methods, showfliers=False)
plt.xticks(rotation=90)
plt.title("Coverage Distribution across folds")
plt.ylabel("Coverage")

plt.subplot(1,2,2)
length_data = [length_results[m] for m in methods]
plt.boxplot(length_data, labels=methods, showfliers=False)
plt.xticks(rotation=90)
plt.title("Interval Length Distribution across folds")
plt.ylabel("Length")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, KFold
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Set random seed
np.random.seed(42)
tf.random.set_seed(42)

# ========== Parameter Settings ==========
alpha = 0.1  # Corresponding to 90% coverage
alpha_l = alpha / 2
alpha_u = 1 - alpha_l

# Load Lending Club dataset
df = pd.read_csv('loan_sampled.csv')

# ========= Data Cleaning ==========
# Drop irrelevant columns
drop_columns = ['id', 'member_id', 'grade', 'issue_d', 'url', 'emp_title']
df.drop(columns=drop_columns, inplace=True, errors='ignore')

# Target variable (loan amount)
target = 'loan_amnt'
y = df[target]
X = df.drop(columns=[target])

# ========= Feature Engineering ==========
# Identify important features(by domain knowledge or feature importance)
important_features = [
    'annual_inc',          # Annual income
    'dti',                 # Debt-to-income ratio
    'emp_length',          # Employment length
    'home_ownership',      # Home ownership status
    'revol_util',
    'purpose',             # Loan purpose
    'sub_grade',           # Loan sub-grade
    'tot_hi_cred_lim',
    'int_rate',            # Interest rate
    'verification_status'  # Loan verification status
]

# Filter important features
X = X[important_features]

# Classify categorical and numerical variables
categorical_features = ['home_ownership', 'sub_grade', 'purpose', 'verification_status']
numerical_features = [col for col in important_features if col not in categorical_features]

# Ensure column names exist in the dataset
categorical_features = [col for col in categorical_features if col in X.columns]
numerical_features = [col for col in numerical_features if col in X.columns]
print("Numerical Features:", numerical_features)
print("Categorical Features:", categorical_features)

# Clean the emp_length column
def clean_emp_length(value):
    if isinstance(value, str):
        value = value.replace('+', '').replace('years', '').replace('year', '').strip()
        if value.isdigit():
            return int(value)
        elif '<' in value:
            return 0
        elif 'n/a' in value.lower():
            return np.nan
    return value

X['emp_length'] = X['emp_length'].apply(clean_emp_length)

# Create preprocessing pipeline: standardize numerical features and one-hot encode categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_features)
    ]
)

pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Preprocess data
X_preprocessed = pipeline.fit_transform(X)
print(type(X_preprocessed), X_preprocessed.shape)
# Get preprocessed feature names
numerical_features_transformed = numerical_features
categorical_features_transformed = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features)
preprocessed_feature_names = list(numerical_features_transformed) + list(categorical_features_transformed)

# Convert preprocessed data to DataFrame
X_preprocessed_df = pd.DataFrame(X_preprocessed, columns=preprocessed_feature_names)
# Use Random Forest to select important features
rf_selector = RandomForestRegressor(n_estimators=100, random_state=42)
rf_selector.fit(X_preprocessed_df, y)

feature_importances = rf_selector.feature_importances_
importance_df = pd.DataFrame({
    'Feature': preprocessed_feature_names,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

print("Feature Importances:")
print(importance_df)

# Select top important features
selector = SelectFromModel(rf_selector, prefit=True, threshold='median')
X_selected = selector.transform(X_preprocessed_df)
selected_features = selector.get_support(indices=True)
selected_feature_names = [preprocessed_feature_names[i] for i in selected_features]
print("Selected Features:", selected_feature_names)

# Update X to only include selected important features
X = pd.DataFrame(X_selected, columns=selected_feature_names)

print(f"Feature selection complete. Shape of X: {X.shape}")


###################################
# Define necessary functions (same as previous code)
###################################
def coverage_and_length(lower_preds, upper_preds, y_true):
    coverage = np.mean((y_true >= lower_preds) & (y_true <= upper_preds))
    avg_length = np.mean(upper_preds - lower_preds)
    return coverage, avg_length

def pinball_loss(alpha_q):
    def loss(y_true, y_pred):
        diff = y_true - y_pred
        return tf.reduce_mean(tf.maximum(alpha_q*diff, (alpha_q-1)*diff))
    return loss

# Mean model training function
def train_ridge(X, y):
    ridge = Ridge(alpha=1.0)
    ridge.fit(X, y)
    return ridge

def train_rf_mean(X, y):
    rf = RandomForestRegressor(n_estimators=1000, random_state=42)
    rf.fit(X, y)
    return rf

def build_nn_mean(input_dim):
    model = models.Sequential([
        layers.Dense(64, activation='relu', input_shape=(input_dim,)),
        layers.Dense(64, activation='relu'),
        layers.Dense(1)
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(5e-4), loss='mse')
    return model

def train_nn_mean(X, y):
    model = build_nn_mean(X.shape[1])
    es = tf.keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)
    model.fit(X, y, epochs=1000, batch_size=64, validation_split=0.1, verbose=0, callbacks=[es])
    return model

# Quantile model training function (GBR)
def train_quantile_gbr(X, y, alpha_q):
    gbr = GradientBoostingRegressor(loss='quantile', alpha=alpha_q,
                                    n_estimators=1000, max_depth=5,
                                    learning_rate=0.05, random_state=42)
    gbr.fit(X, y)
    return gbr

# QNN
def build_nn_quantile(input_dim):
    model = models.Sequential([
        layers.Dense(64, activation='relu', input_shape=(input_dim,)),
        layers.Dense(64, activation='relu'),
        layers.Dense(1)
    ])
    return model

def train_qnn(X, y, alpha_q):
    model = build_nn_quantile(X.shape[1])
    model.compile(optimizer=tf.keras.optimizers.Adam(5e-4), loss=pinball_loss(alpha_q))
    es = tf.keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)
    model.fit(X, y, epochs=1000, batch_size=64, validation_split=0.1, verbose=0, callbacks=[es])
    return model

def train_qnn_pair(X, y, alpha_l, alpha_u):
    model_lower = train_qnn(X, y, alpha_l)
    model_upper = train_qnn(X, y, alpha_u)
    return model_lower, model_upper

# SCP function
def split_conformal_calibrate(f, X_cal, y_cal, alpha=0.1):
    cal_preds = f(X_cal)
    residuals = np.abs(y_cal - cal_preds)
    q_res = np.quantile(residuals, 1 - alpha)
    return q_res

def split_conformal_predict(f, X, q_res):
    preds = f(X)
    return preds - q_res, preds + q_res

# LACP function
def train_mu_sigma_for_lacp(X_train, y_train, mu_model_type='rf'):
    if mu_model_type == 'ridge':
        mu_model = train_ridge(X_train, y_train)
    elif mu_model_type == 'rf':
        mu_model = train_rf_mean(X_train, y_train)
    elif mu_model_type == 'nn':
        mu_model = train_nn_mean(X_train, y_train)
    else:
        raise ValueError("Unknown mu_model_type")

    mu_pred_train = mu_model.predict(X_train)
    mu_pred_train = np.array(mu_pred_train).ravel()
    R_train = np.abs(y_train - mu_pred_train)

    sigma_model = RandomForestRegressor(n_estimators=500, random_state=42)
    sigma_model.fit(X_train, R_train)
    return mu_model, sigma_model

def lacp_calibrate(mu_model, sigma_model, X_cal, y_cal, alpha=0.1):
    mu_pred_cal = mu_model.predict(X_cal)
    sigma_pred_cal = sigma_model.predict(X_cal)
    sigma_pred_cal = np.array(sigma_pred_cal).ravel()
    R_tilde_cal = np.abs(y_cal - mu_pred_cal)/(sigma_pred_cal + 1e-9)
    Q_1_alpha = np.quantile(R_tilde_cal, 1 - alpha)
    return Q_1_alpha

def lacp_predict(mu_model, sigma_model, X, Q_1_alpha):
    mu_pred = mu_model.predict(X)
    sigma_pred = sigma_model.predict(X)
    return mu_pred - sigma_pred*Q_1_alpha, mu_pred + sigma_pred*Q_1_alpha

# CQR function
def cqr_calibrate(q_lower, q_upper, X_cal, y_cal, alpha=0.1):
    cal_lower = q_lower(X_cal)
    cal_upper = q_upper(X_cal)
    E = np.maximum(cal_lower - y_cal, y_cal - cal_upper)
    Q = np.quantile(E, 1 - alpha)
    return Q

def cqr_predict(q_lower, q_upper, X, Q):
    lower_preds = q_lower(X) - Q
    upper_preds = q_upper(X) + Q
    return lower_preds, upper_preds

###################################
# Use KFold for cross-validation
###################################
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# We need to record coverage and length for various methods
methods = [
    'Ridge','Random Forest','Neural Net',
    'Ridge_LACP','RF_LACP','NN_LACP',
    'CQR_QRF','CQR_QNN',
    'QRF_Quantile_NoConformal','QNN_Quantile_NoConformal'
]

coverage_results = {m:[] for m in methods}
length_results = {m:[] for m in methods}

# To display intervals for CQR(GBR) at the last fold
cqr_gbr_interval_lower = None
cqr_gbr_interval_upper = None
y_test_final = None

X_np = X.values if hasattr(X,'values') else X
y_np = y.values if hasattr(y,'values') else y

for fold_id,(train_index, test_index) in enumerate(kf.split(X_np)):
    X_train_full, X_test_fold = X_np[train_index], X_np[test_index]
    y_train_full, y_test_fold = y_np[train_index], y_np[test_index]

    # Split train/calibration sets
    n_cal = int(0.2 * len(X_train_full))
    X_train_fold, X_cal_fold = X_train_full[:-n_cal], X_train_full[-n_cal:]
    y_train_fold, y_cal_fold = y_train_full[:-n_cal], y_train_full[-n_cal:]

    # Train mean models
    ridge = train_ridge(X_train_fold, y_train_fold)
    rf_mean = train_rf_mean(X_train_fold, y_train_fold)
    nn_mean = train_nn_mean(X_train_fold, y_train_fold)

    f_ridge = lambda X: ridge.predict(X)
    f_rf = lambda X: rf_mean.predict(X)
    f_nn = lambda X: nn_mean.predict(X).flatten()

    # Quantile models (GBR Quantile)
    model_lower_gbr = train_quantile_gbr(X_train_fold, y_train_fold, alpha_l)
    model_upper_gbr = train_quantile_gbr(X_train_fold, y_train_fold, alpha_u)
    f_gbr_lower = lambda X: model_lower_gbr.predict(X)
    f_gbr_upper = lambda X: model_upper_gbr.predict(X)

    # QNN
    model_lower_qnn, model_upper_qnn = train_qnn_pair(X_train_fold, y_train_fold, alpha_l, alpha_u)
    f_qnn_lower = lambda X: model_lower_qnn.predict(X).flatten()
    f_qnn_upper = lambda X: model_upper_qnn.predict(X).flatten()

    # ========== SCP ==========
    q_res_ridge_scp = split_conformal_calibrate(f_ridge, X_cal_fold, y_cal_fold, alpha=alpha)
    ridge_lower_scp, ridge_upper_scp = split_conformal_predict(f_ridge, X_test_fold, q_res_ridge_scp)
    cov, length = coverage_and_length(ridge_lower_scp, ridge_upper_scp, y_test_fold)
    coverage_results['Ridge'].append(cov)
    length_results['Ridge'].append(length)

    q_res_rf_scp = split_conformal_calibrate(f_rf, X_cal_fold, y_cal_fold, alpha=alpha)
    rf_lower_scp, rf_upper_scp = split_conformal_predict(f_rf, X_test_fold, q_res_rf_scp)
    cov, length = coverage_and_length(rf_lower_scp, rf_upper_scp, y_test_fold)
    coverage_results['Random Forest'].append(cov)
    length_results['Random Forest'].append(length)

    q_res_nn_scp = split_conformal_calibrate(f_nn, X_cal_fold, y_cal_fold, alpha=alpha)
    nn_lower_scp, nn_upper_scp = split_conformal_predict(f_nn, X_test_fold, q_res_nn_scp)
    cov, length = coverage_and_length(nn_lower_scp, nn_upper_scp, y_test_fold)
    coverage_results['Neural Net'].append(cov)
    length_results['Neural Net'].append(length)

    # ========== LACP ==========
    # Ridge
    mu_ridge, sigma_ridge = train_mu_sigma_for_lacp(X_train_fold, y_train_fold, mu_model_type='ridge')
    Q_ridge_lacp = lacp_calibrate(mu_ridge, sigma_ridge, X_cal_fold, y_cal_fold, alpha=alpha)
    ridge_lower_lacp, ridge_upper_lacp = lacp_predict(mu_ridge, sigma_ridge, X_test_fold, Q_ridge_lacp)
    cov, length = coverage_and_length(ridge_lower_lacp, ridge_upper_lacp, y_test_fold)
    coverage_results['Ridge_LACP'].append(cov)
    length_results['Ridge_LACP'].append(length)

    # RF
    mu_rf, sigma_rf = train_mu_sigma_for_lacp(X_train_fold, y_train_fold, mu_model_type='rf')
    Q_rf_lacp = lacp_calibrate(mu_rf, sigma_rf, X_cal_fold, y_cal_fold, alpha=alpha)
    rf_lower_lacp, rf_upper_lacp = lacp_predict(mu_rf, sigma_rf, X_test_fold, Q_rf_lacp)
    cov, length = coverage_and_length(rf_lower_lacp, rf_upper_lacp, y_test_fold)
    coverage_results['RF_LACP'].append(cov)
    length_results['RF_LACP'].append(length)

    # NN
    mu_nn, sigma_nn = train_mu_sigma_for_lacp(X_train_fold, y_train_fold, mu_model_type='nn')
    Q_nn_lacp = lacp_calibrate(mu_nn, sigma_nn, X_cal_fold, y_cal_fold, alpha=alpha)
    nn_lower_lacp, nn_upper_lacp = lacp_predict(mu_nn, sigma_nn, X_test_fold, Q_nn_lacp)
    cov, length = coverage_and_length(nn_lower_lacp, nn_upper_lacp, y_test_fold)
    coverage_results['NN_LACP'].append(cov)
    length_results['NN_LACP'].append(length)

    # ========== CQR ==========
    # CQR (QRF)
    Q_gbr = cqr_calibrate(f_gbr_lower, f_gbr_upper, X_cal_fold, y_cal_fold, alpha=alpha)
    gbr_lower_cqr, gbr_upper_cqr = cqr_predict(f_gbr_lower, f_gbr_upper, X_test_fold, Q_gbr)
    cov, length = coverage_and_length(gbr_lower_cqr, gbr_upper_cqr, y_test_fold)
    coverage_results['CQR_QRF'].append(cov)
    length_results['CQR_QRF'].append(length)

    # CQR (QNN)
    Q_qnn = cqr_calibrate(f_qnn_lower, f_qnn_upper, X_cal_fold, y_cal_fold, alpha=alpha)
    qnn_lower_cqr, qnn_upper_cqr = cqr_predict(f_qnn_lower, f_qnn_upper, X_test_fold, Q_qnn)
    cov, length = coverage_and_length(qnn_lower_cqr, qnn_upper_cqr, y_test_fold)
    coverage_results['CQR_QNN'].append(cov)
    length_results['CQR_QNN'].append(length)

    # ========== Quantile baseline (No Conformal) ==========
    # QRF
    gbr_lower = f_gbr_lower(X_test_fold)
    gbr_upper = f_gbr_upper(X_test_fold)
    cov, length = coverage_and_length(gbr_lower, gbr_upper, y_test_fold)
    coverage_results['QRF_Quantile_NoConformal'].append(cov)
    length_results['QRF_Quantile_NoConformal'].append(length)

    # QNN
    qnn_lower = f_qnn_lower(X_test_fold)
    qnn_upper = f_qnn_upper(X_test_fold)
    cov, length = coverage_and_length(qnn_lower, qnn_upper, y_test_fold)
    coverage_results['QNN_Quantile_NoConformal'].append(cov)
    length_results['QNN_Quantile_NoConformal'].append(length)

    # Record the CQR (GBR) result of the last fold for visualization
    if fold_id == kf.n_splits - 1:
        cqr_gbr_interval_lower = gbr_lower_cqr
        cqr_gbr_interval_upper = gbr_upper_cqr
        y_test_final = y_test_fold

###################################
# Output average result
###################################
print("=== Average Results over 5 folds ===")
for m in methods:
    avg_cov = np.mean(coverage_results[m])
    avg_len = np.mean(length_results[m])
    print(f"{m}: Coverage={avg_cov:.3f}, Length={avg_len:.3f}")

###################################
# Plotting length and coverage comparison graph
###################################
avg_coverages = [np.mean(coverage_results[m]) for m in methods]
avg_lengths = [np.mean(length_results[m]) for m in methods]

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.barplot(x=methods, y=avg_coverages)
plt.xticks(rotation=90)
plt.title("Average Coverage Comparison")
plt.ylabel("Coverage")

plt.subplot(1,2,2)
sns.barplot(x=methods, y=avg_lengths)
plt.xticks(rotation=90)
plt.title("Average Length Comparison")
plt.ylabel("Length")
plt.tight_layout()
plt.show()

###################################
# Sample plot showing prediction intervals and actual values for CQR (GBR)
###################################
# Only a portion of the samples (such as the first 50 samples) are displayed
num_samples_to_plot = min(50,len(y_test_final))
x_axis = np.arange(num_samples_to_plot)

plt.figure(figsize=(10,6))
plt.plot(x_axis, y_test_final[:num_samples_to_plot], 'o', color='blue', label='True Values')
mid = (cqr_gbr_interval_lower[:num_samples_to_plot] + cqr_gbr_interval_upper[:num_samples_to_plot]) / 2
plt.plot(x_axis, mid, color='green', label='CQR(QRF) Prediction (Mid)')
plt.fill_between(x_axis, cqr_gbr_interval_lower[:num_samples_to_plot],
                 cqr_gbr_interval_upper[:num_samples_to_plot],
                 color='green', alpha=0.2, label='CQR(QRF) Interval')
plt.xlabel('Test Sample Index')
plt.ylabel('House Price')
plt.title('CQR(QRF) Prediction Intervals vs. True Values')
plt.legend()
plt.show()

###################################
# Use boxplot to show the distribution of different methods under K-fold
###################################

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
coverage_data = [coverage_results[m] for m in methods]
plt.boxplot(coverage_data, labels=methods, showfliers=False)
plt.xticks(rotation=90)
plt.title("Coverage Distribution across folds")
plt.ylabel("Coverage")

plt.subplot(1,2,2)
length_data = [length_results[m] for m in methods]
plt.boxplot(length_data, labels=methods, showfliers=False)
plt.xticks(rotation=90)
plt.title("Interval Length Distribution across folds")
plt.ylabel("Length")

plt.tight_layout()
plt.show()
```
